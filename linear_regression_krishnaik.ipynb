linear regression practical (sklearn bostan_dataset)

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as py
```

```python
from sklearn.datasets import load_boston
```

```python
df=load_boston()
df
```

```python
dataset=pd.DataFrame(df.data,columns=df.feature_names)
dataset
```

```python
dataset.head(10)
```

```python
dataset.tail(10)
```

```python
## independent and dependt feature

X = dataset
y= df.target
```

```python
y
```

```python
## train_test_Split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
 y_train
```

```python
## standarised the dataset 

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

##cross validation 

from sklearn.model_selection import cross_val_score
```

```python
regression = LinearRegression()
regression.fit(X_train,y_train)
```

```python

mse=cross_val_score(regression,X_train,y_train,scoring='neg_mean_squared_error',cv=15)
```

```python
np.mean(mse)
```

```python
## prediction 

reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.kdeplot(reg_pred-y_test)

plt.show()
```

```python
from sklearn.metrics import r2_score

score=r2_score(reg_pred,y_test)
score
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
```

```python
ridge_regressor = Ridge()
```

```python
ridge_regressor
```

```python
parameter = {'alpha':[1,2,5,10,30,40,50,60,70,90,80]}

ridgecv=GridSearchCV(ridge_regressor,parameter,scoring='neg_mean_squared_error',cv=5)
ridgecv.fit(X_train,y_train)
```

```python
print(ridgecv.best_params_)
```

```python
print(ridgecv.best_score_)
```

```python
ridge_pred = ridgecv.predict(X_test)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.kdeplot(ridge_pred-y_test)

plt.show()
```
